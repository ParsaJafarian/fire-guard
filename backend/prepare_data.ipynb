#### Download 2023 and 2024 datasets from CWFIS

```python
#Define download function
import requests
def download(base_url: str, filename: str):
  """
  Download a file from the given base url.
  base_url - base url where file is stored. Must end with `/`
  filename - name of the file to download. Must contain suffix (.csv, .json, etc)
  """
  response = requests.get(base_url + filename)

  # Check if the request was successful (status code 200)
  if response.status_code == 200:
      # Write the content to a local file
      with open(filename, 'wb') as f:
          f.write(response.content)
      print('File downloaded successfully')
  else:
      print('Failed to download file')
```

### Adding elevation by using Canada's api


```python
url = "http://geogratis.gc.ca/services/elevation/cdem/altitude?"
def get_altitude(lat: float, lon: float):
  """
  Get the altitude of a geographical point using canada's elevation api:
  http://geogratis.gc.ca/services/elevation/cdem/altitude?

  Args:
      lat (float): Latitude
      lon (float): Longitude

  Returns:
      float: The altitude at the given latitude & longitude
  """

  response = requests.get(f"{url}lat={lat}&lon={lon}")

  # Check if the request was successful (status code 200)
  if response.status_code == 200:
      try:
            # Parse the response JSON
            data = response.json()
            # Extract the altitude value (assuming the key is 'altitude')
            altitude = data.get('altitude')
            if altitude is not None:
                print("Altitude obtained")
                return altitude
            else:
                print('Altitude data not found in response')
      except ValueError:
            print('Failed to parse JSON response')
  else:
      print('Failed to get altitude')

  return None
```

### Merging and Saving Datasets
Merge all the datasets and keep only rows that are after 2000. We do this because the climate has changes so much and old data might confuse the model. Moreover, having too much data is computationally expensive for other data manipulations.

```python
df = df.drop("biome", axis=1)
df.head(1)
```

```python
# Concatenate datasets
df_merged = pd.concat([df, df_2023, df_2024])

# Ensure 'date' column is in datetime format
df_merged['date'] = pd.to_datetime(df_merged['date'], errors='coerce')

# Drop rows where 'date' is NaT (Not a Time)
df_merged = df_merged.dropna(subset=['date'])

# Extract the year from the 'date' column
df_merged['year'] = df_merged['date'].dt.year

# Filter rows where the year is greater than or equal to 2000
df_merged = df_merged[df_merged['year'] >= 2000]

# Drop the temporary 'year' column
df_merged = df_merged.drop('year', axis=1)

# Check the number of remaining rows
print(len(df_merged))
```
