# Exploratory Data Analysis
This notebook performs EDA on `clean-weather.csv`, which has now been preprocessed and cleaned largely by now

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean-weather.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
# Bar plot Count vs. Year
year_counts = df['LOCAL_YEAR'].value_counts().sort_index()

# Plot the counts
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar')
plt.title('Row Count by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot Count vs. Province
year_counts = df['PROVINCE_CODE'].value_counts().sort_index()

# Plot the counts
plt.figure(figsize=(10, 6))
year_counts.plot(kind='bar')
plt.title('Row Count by Province')
plt.xlabel('Province')
plt.ylabel('Count')
plt.show()
```
